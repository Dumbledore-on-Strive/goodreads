```
#visualization of average rating by length of book


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('df_goodreads.csv', error_bad_lines=False)
    
sns.jointplot(x=df["avg_rating"], y=df["num_pages"], kind='hex', marginal_kws=dict(bins=30, fill=True))

plt.show()
```

```
#visualisation of the minmax_norm_rating distribution

df = pd.read_csv('df_goodreads.csv', error_bad_lines=False)
df['minmax_norm_ranges'] = np.around(MinMaxScaler((1, 10)).fit_transform(df[['avg_rating']]), 2)
df['minmax_norm_ranges']

sns.kdeplot(df['minmax_norm_ranges'])
```

```
#visualisation ofthe mean_norm_rating distribution
df['mean_norm_ratings'] = np.around((1 + (df['avg_rating'] - df['avg_rating'].mean()) / (df['avg_rating'].max() - df['avg_rating'].min()) * 9), 2)
sns.kdeplot(df['mean_norm_ratings'], color="#FFC0CB")
```
